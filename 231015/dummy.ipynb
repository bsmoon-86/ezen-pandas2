```python
import random
import string
```

```python
# 문자열 랜덤 값을 생성
letter_set = string.ascii_lowercase
letter_set
```

```python
random_data = random.sample(letter_set, 5)
# list를 문자열로 변경
data = "".join(random_data)
data
```

```python
text_length = range(4, 10)
email = ['@naver.com', '@gmail.com', '@hanmail.net', '@yahoo.com']

email_list = []
for i in range(1000000):
    # email_list의 길이가 100일때 반복문을 종료
    if (len(email_list) == 100):
        break

    random_data = random.sample(letter_set, random.choice(text_length))

    data = "".join(random_data) + random.choice(email)

    # data를 email_list에 대입 -> 해당하는 이메일이 존재하지 않는 경우에만
    if (data not in email_list):
        email_list.append(data)
    
```

```python
from datetime import datetime
```

```python
year = list(range(2017, 2024))
month = list(range(1, 13))
day = list(range(1, 32))

time1 = []
for i in range(100000):
    y = str(random.choice(year))
    m = str(random.choice(month))
    d = str(random.choice(day))

    date = y+m+d

    try:
        time1.append(datetime.strptime(date, '%Y%m%d'))
    except:
        print(date)
    
    if (len(time1) == 100):
        break
```

```python
time1
```

```python
year = list(range(2017, 2024))
month = list(range(1, 13))
day = list(range(1, 32))

time2 = []

# time1의 인덱스 값 설정
time1_index = 0
for i in range(100000):
    y = str(random.choice(year))
    m = str(random.choice(month))
    d = str(random.choice(day))

    date = y+m+d

    try:
        a = datetime.strptime(date, '%Y%m%d')
        if (time1[time1_index] < a):
            print(time1[time1_index] < a)
            print(time1_index)
            time2.append(a)
            time1_index = time1_index + 1
    except:
        print(date)
    
    if (len(time2) == 100):
        break
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'email' : email_list, 
    '입사일' : time1, 
    '퇴사일' : time2
})
```

```python
flag = df['입사일'] < '20171231'
flag2 = df['퇴사일'] > '20171231'
```

```python
def func_1(_df, _date):
    _date1 = _date+"1231"
    _date2 = _date+"0101"
    flag = _df['입사일'] < _date1
    flag2 = _df['퇴사일'] > _date2

    return _df.loc[flag & flag2]
```

```python
func_1(df, '2023')
```

```python
for year in range(2017, 2024):
    data = func_1(df, str(year))
    data.to_excel(f"./excel_data/{year}.xlsx")
```

```python
import calendar
```

```python
def func_2(_df, _year, _month):
    print(len(str(_month)))
    if(len(str(_month)) == 1):
        _date = str(_year) + '0' +str(_month)
    else:
        _date = str(_year) +str(_month)
    _date1 = _date + str(calendar.monthrange(_year, _month)[-1])
    _date2 = _date + '0' +str(calendar.monthrange(_year, _month)[0])
    print(_date,_date1, _date2)
    flag = df['입사일'] < _date1
    flag2 = df['퇴사일']  > _date2
    return _df.loc[flag & flag2]
```

```python
func_2(df, 2017, 2)
```

```python
def func_3(_df, _year, _month):
    last_day = calendar.monthrange(_year, _month)[-1]
    first_day = calendar.monthrange(_year, _month)[0]
    _date1 = datetime(_year, _month, last_day)
    _date2 = datetime(_year, _month, first_day+1)
    print(_date1, _date2)
    flag = df['입사일'] < _date1
    flag2 = df['퇴사일']  > _date2
    return _df.loc[flag & flag2]
```

```python
calendar.monthrange(2017, 5)
```

```python
func_3(df, 2017, 5)
```

```python
# 다중 for문 사용
# 첫번째 for문 (2017 - 2023)
for year in range(2017, 2024):
    # 두번째 for문 (1 - 12)
    for month in range(1, 13):
        # print(year, month)
        data = func_3(df, year, month)
        data.to_excel(f"./excel_data/{year}-{month}.xlsx")
```
